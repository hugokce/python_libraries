```python
#using the map method on the 'Age' column  
dftrain['old_or_not'] = dftrain['Age'].map(lambda age: 1 if age >=45 else 0)
```

```python
dftrain
```

```python
#using the map method on the 'Sex' column  
dftrain['boolean_sex'] = dftrain['Sex'].map({'male' : 0, 'female': 1})
```

```python
dftrain
```

```python
#define a function that converts dollars to euro
def dolar_to_euro(dollars, exchange_rate):
    euros = dollars*exchange_rate
    return euros
```

```python
#define a function that converts dollars to euro
def dolar_to_tl(dollars, exchange_rate):
    tl = dollars*exchange_rate
    return tl
```

```python
#create a new column called Fare_euros
dftrain['Fare_euros'] = dftrain['Fare'].apply(dolar_to_euro,exchange_rate = 0.9)
```

```python
dftrain
```

```python
#create a new column called old_woman
dftrain['old_woman'] = dftrain.apply(lambda row: row['boolean_sex'] and row['old_or_not'], axis = 1)
```
