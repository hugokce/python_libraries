```python
import pandas as pd
```

```python
df=pd.read_csv('sample.csv',sep=',',header=0)
df.head()
```

```python
def cleanup_salary(row):
    salary=row['Salary'].replace('$','')
    salary=float(salary)
    return salary
```

```python
df['clean_salary']=df.apply(cleanup_salary,axis=1)
```

```python
print('Average Salary Per Gender')
print(df.groupby('gender')['clean_salary'].mean())
```

```python
def female_salary(row):
    if row['gender']=='Female':
        return row['clean_salary']
```

```python
df['female_salary']=df.apply(female_salary,axis=1)
```

```python
def male_salary(row):
    if row['gender']=='Male':
        return row['clean_salary']
```

```python
df['male_salary']=df.apply(male_salary,axis=1)
```

```python
df=pd.read_csv('exercise_course.csv',sep=',',header=0)
```

```python
df2=df.groupby('url')['bytes'].sum().reset_index()
```

```python
df2=df2.sort_values('bytes',ascending=False)
print(df2)
```

```python
def female_bytes(row):
    if row['gender']=='Female':
        return row['bytes']
    
def male_bytes(row):
    if row['gender']=='Male':
        return row['bytes']
```

```python
df['female_bytes']=df.apply(female_bytes,axis=1)
df['male_bytes']=df.apply(male_bytes,axis=1)
```

```python
df3=df.groupby('url')['male_bytes','female_bytes'].sum()
print(df3)
```

```python
df4=df.groupby(['url','country'])['male_bytes','female_bytes'].sum()
print(df4)
```

```python
df4=df.groupby(['country','url'])['male_bytes','female_bytes'].sum()
print(df4)
```
