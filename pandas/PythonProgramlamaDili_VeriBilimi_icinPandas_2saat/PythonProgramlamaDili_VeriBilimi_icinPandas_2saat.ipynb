```python
#GroupBy metodu
sozluk = {"Fakülte":["Makine M.","Makine M.","Endüstri M.",
                     "Endüstri M.","Mekatronik M."],
         "Öğrenci Sayısı":[1500,1200,500,250,300]}
```

```python
tablo=pd.DataFrame(sozluk)
tablo
```

```python
fakülte = tablo.groupby("Fakülte")
fakülte.count()
```

```python
fakülte.mean()
```

```python
fakülte.min()
```

```python
fakülte.max()
```

```python
fakülte.describe()
```

```python
#Concat işlemi
sozluk1 = {"A":["a","b","c","d"],
         "B":["e","f","g","h"],
         "C":[10,20,30,40]}
```

```python
tablo1 = pd.DataFrame(sozluk1,index=range(4))
tablo1
```

```python
sozluk2 = {"A":["i","j","k","l"],
         "B":["m","n","o","p"],
         "C":[50,60,70,80]}
```

```python
tablo2 = pd.DataFrame(sozluk2,index=[4,5,6,7])
tablo2
```

```python
sozluk3 = {"A":["q","r","s","t"],
         "B":["u","v","w","x"],
         "C":[90,100,110,120]}
```

```python
tablo3 = pd.DataFrame(sozluk3,index=[8,9,10,11])
tablo3
```

```python
pd.concat([tablo1,tablo2,tablo3])
```

```python
#merge komutu
sozluk1 = {"A":["a","b","c","d"],
         "B":["x","y","z","t"]
         }
```

```python
tablo_m=pd.merge(tablo1,tablo2,on="A")
tablo_m
```
