Fill Plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(0,10)
Y = 0.05*X*X
```

```python
plt.ylim(0,5)
plt.plot(X,Y)
plt.fill_between(X,Y,color='blue',alpha='0.5')  #bu olmazsa düz çizgi çıkar. 
plt.show()
```

```python
plt.ylim(0,5)
plt.plot(X,Y)
plt.fill_between(X,Y,color='blue',alpha='1')  #bu olmazsa düz çizgi çıkar. alpha 1 olursa mavi renk daha koyu olur.
plt.show()
```

```python
X = np.arange(0,10)
Y1 = 0.05*X*X
Y2 = 0.03*X*X
plt.ylim(0,5)
plt.plot(X,Y1,color ='blue')
plt.plot(X,Y2,color ='red')
plt.fill_between(X,Y1,Y2,color='red',alpha='0.5')  #bu olmazsa düz çizgi çıkar. 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

x = [1,2,3,4,5,6,7]
y = [1,1,1,1,1,1,1]
z = [2,4,6,8,10,12,14]

x2 = [1,2,3,4,5,6,7]
y2 = [1,1,1,1,1,1,1]
z2 = [1,3,5,7,9,11,13]

ax.plot(x,y,z)
ax.plot(x2,y2,z2,color='red')
plt.show()
```
