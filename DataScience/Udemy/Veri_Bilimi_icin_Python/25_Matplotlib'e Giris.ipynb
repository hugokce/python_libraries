#### Stilleri Ayarlama

Figürlerimiz için uygun estetik tarzları seçmek için **plt.style**  kullanacağız. 

```python
import numpy as np
deger = np.arange(0,10)
```

```python
plt.style.use('classic')
plt.plot(deger)
plt.show()
```

```python
plt.style.use('dark_background')
plt.plot(deger)
plt.show()
```

#### Görseli kaydetmek

```python
plt.style.use('classic')#stil değiştirdik

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.show()
```

```python
fig.savefig('first_figure.png')
```
