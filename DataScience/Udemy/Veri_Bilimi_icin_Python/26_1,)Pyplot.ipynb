# Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x_dizi=np.arange(1,5,1)
y_dizi=np.arange(1,5,1)

print('x eksenindeki koor. : ',x_dizi)
print('y eksenindeki koor. : ',y_dizi)
```

```python
plt.figure() 

plt.plot(x_dizi,y_dizi, 'ro')

plt.show()
```

**Değerleri daha iyi gözlemlemek için plt.grid(True) yapmalıyız.**

```python
plt.figure() 
plt.plot(x_dizi,y_dizi, 'ro')
plt.grid(True)
plt.show()
```

```python
plt.figure() 
plt.plot(x_dizi,y_dizi,'^g')
plt.grid(True)
plt.show()
```

```python
plt.figure() 

plt.title('X-Y Ekseni')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')


plt.plot(x_dizi,y_dizi,'bs')
plt.grid(True)
plt.show()
```

#### Çizimin stilini biçimlendirmek

```python
x_dizi_kare = x_dizi
y_dizi_kare =y_dizi** 2

print('x_dizi_kare eksenindeki koor. : ',x_dizi_kare)
print('y_dizi_kare eksenindeki koor. : ',y_dizi_kare)
```

```python
plt.figure() 
plt.plot(x_dizi_kare,y_dizi_kare, 'ro')
plt.grid(True)
plt.show()
```

```python
x_dizi_küp = x_dizi
y_dizi_küp =y_dizi** 3

print('x_dizi_kare eksenindeki koor. : ',x_dizi_küp)
print('y_dizi_kare eksenindeki koor. : ',y_dizi_küp)
```

```python
plt.figure() 
plt.plot(x_dizi_küp,y_dizi_küp, '^g')
plt.grid(True)
plt.show()
```

```python
plt.figure() 
plt.plot(x_dizi,y_dizi, 'bs')# önceki çizimi hatırlayılım
plt.grid(True)
plt.show()
```

## Bu 3 çizimi birleştirelim.

```python
plt.figure() 
plt.plot(x_dizi,y_dizi, '-bs',x_dizi_kare,y_dizi_kare, '-ro',x_dizi_küp,y_dizi_küp, '-^g')
plt.grid(True)
plt.show()
```
