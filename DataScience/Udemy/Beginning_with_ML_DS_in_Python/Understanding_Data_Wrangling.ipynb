Cleaning up messy data

```python
requests = pd.read_csv('311-service-requests.csv')
```

```python
#unique metodu ile datayı inceleriz. Ona göre NA, sabit değer vs için karar alıp datayı ona göre düzenleriz.
#Incident Zip kolonunu inceleyelim
requests['Incident Zip'].unique()
#nan, 83.0, 55164-0737 gibi degerler oldugunu gorduk.
```

```python
#Fixing the nan values and string/float confusion
#NA olabilecek degerleri bir degiskene atayalım
na_values = ['NO CLUE','N/A','0']
requests = pd.read_csv('311-service-requests.csv',na_values=na_values,dtype={'Incident Zip':str})
#Incident Zip kolonunun datatypeını string yaptık. Daha önce float idi.
```

```python
requests['Incident Zip'].unique()
```

```python
#Alt çizgili degerler ile ilgilenelim. 29616-0759', '35209-3114' gibi
rows_with_dashes = requests['Incident Zip'].str.contains('-').fillna(False)
len(requests[rows_with_dashes])
```

```python
requests[rows_with_dashes]
# Bu - 5 degeri gorelim
```

```python
#Bu 5 adet 5digitten büyük satırın degerlerini bir degiskene atayalım
long_zip_codes = requests['Incident Zip'].str.len()>5
requests['Incident Zip'][long_zip_codes].unique()
```

```python
requests['Incident Zip'] = requests['Incident Zip'].str.slice(0,5)
#Ilk 5 karakteri aldık.
```

```python
requests[requests['Incident Zip'] =='00000']
# 2 kaydın zip kodu 00000 bunlara bakıyoruz şimdi. Bu degerlere nan atayacagız
```

```python
zero_zips = requests['Incident Zip'] == '00000'
requests.loc[zero_zips,'Incident Zip'] = np.nan
```

```python
unique_zips = requests['Incident Zip'].unique()
#unique_zips.sort()
unique_zips
```

```python
zips = requests['Incident Zip']
is_close = zips.str.startswith('0') | zips.str.startswith('1')
is_far = ~(is_close) & zips.notnull()
```

```python
zips[is_far]
```

Parsing Unix timestamps

```python
#Unix dosyasını okuyup son satırını sileceğiz.
popcon = pd.read_csv('popularity-contest.tsv',sep=' ',)[:-1]
popcon.columns = ['atime','ctime','package-name','mru-program','tag']
#The colums are the access time, created time, package name, recently used program, and a tag
```

```python
popcon[:5]
```

```python
popcon['atime'] = popcon['atime'].astype(int)
popcon['ctime'] = popcon['ctime'].astype(int)
```

```python
#We can use the pd.to_datetime function to convert our integer timestamps into datetimes.
popcon['atime'] = pd.to_datetime(popcon['atime'], unit='s')
popcon['ctime'] = pd.to_datetime(popcon['ctime'], unit='s')
```

```python
#If we look at the dtype now, it's <M8[ns]. As far as I can tell M8 is secret code for datetime64.
popcon['atime'].dtype
```

```python
popcon[:5]
```

```python
popcon = popcon[popcon['atime'] > '1970-01-01']
```

```python
nonlibraries = popcon[~popcon['package-name'].str.contains('lib')]
nonlibraries.sort_values(by='ctime',ascending=False)[:10]
```

Reading data from SQL databases

```python
import pandas as pd
import sqlite3
```

So far we've only talked about reading data from CSV files. That's a pretty common way to store data, but there are many others! Pandas can read from HTML, JSON, SQL, Excel (!!!), HDF5, Stata, and a few other things. In this chapter we'll talk about reading data from SQL databases.

You can read data from a SQL database using the pd.read_sql function. read_sql will automatically convert SQL column names to DataFrame column names.

read_sql takes 2 arguments: a SELECT statement, and a database connection object. This is great because it means you can read from any kind of SQL database -- it doesn't matter if it's MySQL, SQLite, PostgreSQL, or something else.

```python
con = sqlite3.connect("weather-2012-sqlite/weather_2012.sqlite")
df = pd.read_sql("SELECT * FROM weather_2012 LIMIT 5",con)
df
```

```python
#id kolonunu index olarak tanımlayalım
df = pd.read_sql("SELECT * FROM weather_2012 limit 5 ", con,index_col='id')
df
```

```python
#birden fazla kolonu indexlemek istersek
df = pd.read_sql("SELECT * FROM weather_2012 limit 5", con, index_col=['id','date_time'])
df
```

Writing to a SQLite database

You'll notice that this function is in pd.io.sql. There are a ton of useful functions for reading and writing various kind of data in pd.io, and it's worth spending some time exploring them. (see the documentation!)
http://pandas.pydata.org/pandas-docs/stable/io.html

```python
weather_df = pd.read_csv('weather-2012.csv')
con = sqlite3.connect("test_db.sqlite")
con.execute("DROP TABLE IF EXISTS weather_2012")
weather_df.to_sql("weather_2012",con)
```

```python
#We can now read from the weather_2012 table in test_db.sqlite, and we see that we get the same data back
con = sqlite3.connect("test_db.sqlite")
df = pd.read_sql("SELECT * FROM weather_2012 limit 5 ", con)
df
```

```python
con = sqlite3.connect("test_db.sqlite")
df = pd.read_sql("SELECT * FROM weather_2012 order by weather limit 5 ", con)
df
```

Connecting to other kinds of database

If you have a PostgreSQL database or MySQL database, reading from it works exactly the same way as reading from a SQLite database. You create a connection using psycopg2.connect() or MySQLdb.connect(), and then use
