```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
type(tips)
```

```python
#bar plot
plt.bar(x=tips['smoker'],height=tips['tip'],data=tips)
plt.xlabel('smoker')
plt.ylabel('total_bill')
plt.title('Tips/Smoker')
plt.show()
```

```python
sns.barplot(x='smoker',y='tip',data=tips)
plt.title('Tips vs Smoker')
plt.show()
```

```python
fig1,axes1=plt.subplots()
sns.barplot(x='smoker',y='total_bill',data=tips,hue='sex',
           estimator=np.sum,errcolor='r',errwidth=0.75,
           capsize=0.2,ax=axes1)
axes1.set_xlabel('Smoker - Yes/No')
axes1.set_ylabel('Bill Amount')
axes1.set_title('Total Bill vs Smoker')
plt.show()
```
